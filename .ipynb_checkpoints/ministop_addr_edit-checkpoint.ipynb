# 미니스탑 매장 주소 정리

```python
import pandas as pd

mini_seoul = pd.read_csv('./data/ministop_store_서울.csv', encoding='cp949', index_col=0, header=0, engine='python')

mini_seoul.head()
```

```python
addr = []
for address in mini_seoul.address:
    addr.append(str(address).split())
    
print(len(addr))
```

```python
addr2 = []

for i in range(len(addr)):
    if addr[i][0] == "서울":
        addr[i][0] = "서울특별시"
    if addr[i][0] == "서울시":
        addr[i][0] = "서울특별시"
        
    addr2.append(' '.join(addr[i]))

addr2
```

```python
addr2 = pd.DataFrame(addr2, columns=['address2'])
addr2
```

```python
mini_seoul2 = pd.concat([mini_seoul, addr2], axis=1)
mini_seoul2.tail()
```

```python
gu_list = []

for addr in mini_seoul2.address2 :
    if '강남구' in addr:
        gu='강남구'
    elif '강동구' in addr:
        gu='강동구'

    elif '강북구' in addr:
        gu='강북구'

    elif '강서구' in addr:
        gu='강서구'

    elif '관악구' in addr:
        gu='관악구'

    elif '광진구' in addr:
        gu='광진구'

    elif '구로구' in addr:
        gu='구로구'

    elif '금천구' in addr:
        gu='금천구'

    elif '노원구' in addr:
        gu='노원구'

    elif '도봉구' in addr:
        gu='도봉구'

    elif '동대문구' in addr:
        gu='동대문구'

    elif '동작구' in addr:
        gu='동작구'

    elif '마포구' in addr:
        gu='마포구'

    elif '서대문구' in addr:
        gu='서대문구'

    elif '서초구' in addr:
        gu='서초구'

    elif '성동구' in addr:
        gu='성동구'

    elif '성북구' in addr:
        gu='성북구'

    elif '송파구' in addr:
        gu='송파구'

    elif '양천구' in addr:
        gu='양천구'

    elif '영등포구' in addr:
        gu='영등포구'

    elif '용산구' in addr:
        gu='용산구'

    elif '은평구' in addr:
        gu='은평구'

    elif '종로구' in addr:
        gu='종로구'

    elif '중구' in addr:
        gu='중구'

    elif '중랑구' in addr:
        gu='중랑구'

    gu_list.append(gu)
```

```python
gu_list = pd.DataFrame(gu_list, columns=['지역구'])
gu_list.tail()
```

```python
ministop_seoul_final = pd.concat([mini_seoul2, gu_list], axis=1)
ministop_seoul_final.tail()
```

```python
ministop_seoul_final.to_csv("./data/mini_seoul_final.csv", encoding='cp949', index=False)
```
